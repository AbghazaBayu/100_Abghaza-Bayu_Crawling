```python
import time 
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import requests
```

```python
url = "https://www.tokopedia.com/search?navsource=&page7=&q=justin%20bieber&srp_component_id=01.07.00.00&srp_page_id=&srp_page_title=&st=product"
```

```python
driver = webdriver.Chrome()
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
soup
```

```python
tabel1= soup.findAll('div',class_='prd_link-product-name css-3um8ox')
```

```python
tabel1= soup.findAll('div',class_='prd_link-product-name css-3um8ox')
data_tabel1 = []
for each_tabel1 in tabel1:
    data_tabel1.append(each_tabel1.text)
```
